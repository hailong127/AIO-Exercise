**10. Dealing with missing values - Filling**

i vi phng n th gi tr mi vo cc  trng, ta
c th s dng cc gi tr mean, median... ca ct d liu tng ng  thay th (vic chn gi
tr  thay th cn ty thuc vo tnh cht ca b d liu, bi ton ang gii quyt...). V d, c
mt vi hng c Revenue mang gi tr null, ta c th gn cho n gi tr trung bnh nh sau:

```python
revenue_mean = data_indexed['Revenue (Millions)'].mean()
print("The mean revenue is: ", revenue_mean)

# We can fill the null values with this mean revenue
data_indexed['Revenue (Millions)'].fillna(revenue_mean, inplace=True)
```

**11. apply() function**

Apply functions c dng khi ta mun thc thi mt hm no  ln cc
hng trong bng d liu. Sau khi thc thi, kt qu tr v t hm chnh l gi tr mi ca hng
tng ng. V d, ta mun phn loi phim theo ba mc  [Good, Average, Bad] da trn
Rating, ta c th nh ngha mt hm  lm u ny v apply n ln DataFrame:

```python
# Classify movies based on ratings:
def rating_group(rating):
    if rating >= 7.5:
        return 'Good'
    elif rating >= 6.0:
        return 'Average'
    else:
        return 'Bad'

# Lets apply this function on our movies data
# creating a new variable in the dataset to hold the rating category
data['Rating_category'] = data['Rating'].apply(rating_group)

data[['Title', 'Director', 'Rating', 'Rating_category']].head()
```
